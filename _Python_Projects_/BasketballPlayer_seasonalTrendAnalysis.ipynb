```python
# Creating Function to avoid repeating and Plot multiple graphs at once -----------------------------
```

```python
def myplot(matrixName, playerList=Players):
    for name in playerList:
        plt.plot(matrixName[Player_dictionary[name]], c=Color_dictionary[name], ls='--', marker='D', label=name)
    
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons, rotation ='vertical')
    plt.show()
```

```python
myplot(Salary)
```
